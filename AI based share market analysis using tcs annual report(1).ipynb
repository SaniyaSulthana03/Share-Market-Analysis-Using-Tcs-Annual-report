```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")    # sentence transformer will build  the models 
# all-MiniLM-L6-v2 model contains the instructions or rules and regulations to build the model(LLM)
# model is the name of the model or variable name which is pointing to vestor model
```

```python
# These are the english statements we are giving to vestor model later we can ask the question it will givw the answer
sentences = [
    "That is a happy person",
    "That is a happy dog",
    "That is a very happy person",
    "Today is a sunny day"
]
```

```python
# encode : we are passing english sentences to Large Language vector modelto generate vectors
# the vectors are catching into the embeddings
embeddings = model.encode(sentences)
#here they are generating the vectors here it will give the four vectors or embedings
```

```python
similarities = model.similarity(embeddings, embeddings)   #rows and columns
print(similarities.shape)                           #semantic search matrix(SSM)
```

## Step3 : Creating the FAISS DataBase

```python
'''
While creating FAISS database we need to pass below information
384 the size of the vector
    1.We need to tell EMBEDDING MODEL we are used
    2.we have to pass our VECTOR SIZE created by Embeddding model
    3.Where to store database
    4.dummy parameter
'''
```

```python
# pip install -qU faiss-cpu
```

```python
# import faiss           # This calculate the vector size
# from langchain_community.docstore.in_memory import InMemoryDocstore   #to store vector db temporarily
# from langchain_community.vectorstores import FAISS   #actual database creation
# from langchain_community.embeddings import HuggingFaceEmbeddings


import faiss
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore
from langchain_core.documents import Document

```

```python
# embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Get vector size (this is how you correctly use embed_query)
vector_size = len(embedding_model.embed_query("hello world"))
index = faiss.IndexFlatL2(vector_size)
```

```python
# # index = faiss.IndexFlatL2(len(embeddings.embed_query("hello world")))   #this gives the size of the vector
# # vector_size = len(embeddings.embed_query("hello world"))
# # below statement creates VECTOR DATABASE
# vector_store = FAISS(
#     embedding_function=embeddings,
#     index=index,
#     docstore=InMemoryDocstore(),
#     index_to_docstore_id={},
# )

vector_store = FAISS(
    embedding_function=embedding_model,  
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

## Step4: Convert Chunks or Documents into vectors and store itin database

```python
vector_store.add_documents(documents=documents)
```
